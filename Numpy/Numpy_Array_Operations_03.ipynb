**Advance Array Manipulation**

```python
# Stacking Array
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b))
```

```python
np.column_stack((a,b))
```

```python
# Splitting Array
c = np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)
```

```python
a = np.vsplit(c,2)
```

```python
for i in a:
  print(i)
```
