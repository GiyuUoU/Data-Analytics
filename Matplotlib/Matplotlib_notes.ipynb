Matplotlib is Pythonâ€™s foundational data visualization library.
It lets you turn raw data into visual insights â€” trends, patterns, comparisons.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Day": np.arange(1, 11),
    "Sales": [120,150,170,160,180,220,240,230,260,300],
    "Orders": [20,25,30,28,35,45,50,48,55,60],
    "Profit": [30,40,45,42,50,70,80,75,90,110]
})
df
```

# **1ï¸âƒ£ LINE PLOT (plt.plot())**
**ðŸ”¹ What it is** -> Shows change over time or sequence.

**ðŸ”¹ When to use** -> Sales trend , Website traffic over days , Stock prices , Any time-based data

```python
'''
--Syntax--
plt.plot(x, y, color, linestyle, marker, label)

| Parameter   | Meaning                            |
| ----------- | ---------------------------------- |
|   x         | Independent variable (time, index) |
|   y         | Dependent variable (metric value)  |
|   color     | Line color                         |
|   linestyle | Solid, dashed, dotted              |
|   marker    | Points on line                     |
|   label     | Used for legend                    |

'''
plt.plot(df["Day"], df["Sales"],color="blue",linestyle="-",marker="o",label="Sales")
plt.title("Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
# Variation 2: With Style Parameters
plt.plot(df["Day"], df["Sales"],
         color="blue",
         linestyle="--",
         linewidth=2,
         marker="o")
plt.show()
```

```python
# Variation 3: Multiple Lines
plt.plot(df["Day"], df["Sales"], label="Sales")
plt.plot(df["Day"], df["Profit"], label="Profit")
plt.legend()
plt.show()
```

```python
# Variation 4: Object-Oriented
fig, ax = plt.subplots()
ax.plot(df["Day"], df["Sales"], color="green")
ax.set_title("Sales Trend")
ax.set_xlabel("Day")
ax.set_ylabel("Sales")
plt.show()
```

# **2ï¸âƒ£ BAR CHART (plt.bar())**
**ðŸ”¹ What it is**  -> Compares discrete categories

**ðŸ”¹ When to use**  -> Sales per category , Orders per day , Revenue by region

```python
'''
---Syntax---
plt.bar(x, height, color, width)

| Parameter | Meaning       |
| --------- | ------------- |
|  x        | Categories    |
|  height   | Values        |
|  color    | Bar color     |
|  width    | Bar thickness |

'''
# Vertical Bar
plt.bar(df["Day"], df["Orders"], color="green")
plt.title("Orders per Day")
plt.xlabel("Day")
plt.ylabel("Orders")
plt.show()
```

```python
# Horizontal Bar
plt.barh(df["Day"], df["Orders"])
plt.show()
```

```python
#  Bar with Width & Color
plt.bar(df["Day"], df["Orders"],
        width=0.6,
        color="orange",
        edgecolor="black")
plt.show()
```

```python
# Grouped Bar Chart
x = np.arange(len(df["Day"]))

plt.bar(x - 0.2, df["Sales"], width=0.4, label="Sales")
plt.bar(x + 0.2, df["Profit"], width=0.4, label="Profit")

plt.xticks(x, df["Day"])
plt.legend()
plt.show()
```

# **3ï¸âƒ£ HISTOGRAM (plt.hist())**
**ðŸ”¹ What it is**  -> Shows distribution of data

**ðŸ”¹ When to use**  -> Salary distribution , Sales distribution , Exam scores

```python
'''
---Syntax---
plt.hist(data, bins, color, edgecolor)

| Parameter   | Meaning          |
| ----------- | ---------------- |
| `data`      | Numeric data     |
| `bins`      | Number of ranges |
| `color`     | Bar color        |
| `edgecolor` | Border color     |

'''
plt.hist(df["Sales"], bins=5, color="orange", edgecolor="black")
plt.title("Sales Distribution")
plt.xlabel("Sales Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# Controlled Bins + Style
plt.hist(df["Sales"],
         bins=6,
         color="purple",
         edgecolor="black",
         alpha=0.7)
plt.show()
```

```python
# Density Histogram
plt.hist(df["Sales"], bins=6, density=True)
plt.show()
```

```python
# Multiple Histograms
plt.hist(df["Sales"], alpha=0.6, label="Sales")
plt.hist(df["Profit"], alpha=0.6, label="Profit")
plt.legend()
plt.show()
```

# **4ï¸âƒ£ SCATTER PLOT (plt.scatter())**
**ðŸ”¹ What it is**  -> Shows relationship between two variables

**ðŸ”¹ When to use**  -> Correlation check , Before ML modeling , Feature relationships

```python
'''
---Syntax---
plt.scatter(x, y, color, s)

| Parameter | Meaning   |
| --------- | --------- |
| `x`       | Feature 1 |
| `y`       | Feature 2 |
| `color`   | Dot color |
| `s`       | Dot size  |

'''
plt.scatter(df["Orders"], df["Sales"], color="red", s=80)
plt.title("Orders vs Sales")
plt.xlabel("Orders")
plt.ylabel("Sales")
plt.show()
```

```python
# Size & Color Control
plt.scatter(df["Orders"], df["Sales"],
            s=100,
            c=df["Profit"],
            cmap="viridis")
plt.colorbar(label="Profit")
plt.show()
```

```python
# Marker Style
plt.scatter(df["Orders"], df["Sales"],
            marker="^",
            edgecolors="black")
plt.show()
```

```python
# Object-Oriented Scatter
fig, ax = plt.subplots()
ax.scatter(df["Orders"], df["Sales"])
ax.set_title("Orders vs Sales")
plt.show()
```

# **5ï¸âƒ£ BOX PLOT (plt.boxplot())**
**ðŸ”¹ What it is**  -> Shows distribution + outliers

**ðŸ”¹ When to use**  -> Salary analysis , Fraud detection , Profit variation

```python
'''
---Syntax---
plt.boxplot(data)

| Part        | Meaning      |
| ----------- | ------------ |
| Middle line | Median       |
| Box         | IQR (spread) |
| Whiskers    | Normal range |
| Dots        | Outliers     |

'''
plt.boxplot(df["Profit"])
plt.title("Profit Distribution")
plt.show()
```

```python
# Multiple Boxes
plt.boxplot([df["Sales"], df["Profit"]],
            labels=["Sales", "Profit"])
plt.show()
```

```python
# Styling Boxplot
plt.boxplot(df["Profit"],
            patch_artist=True,
            boxprops=dict(facecolor="lightblue"))
plt.show()
```

# **6ï¸âƒ£ PIE CHART (plt.pie())**
**ðŸ”¹ What it is**  -> Shows proportion

**ðŸ”¹ When to use**  -> Market share , Category contribution

âš ï¸ Less preferred in analytics â€” but still asked.

```python
'''
---Syntax---
plt.pie(values, labels, autopct)

'''
plt.pie(df["Orders"][:4],labels=["D1","D2","D3","D4"],autopct="%1.1f%%")
plt.title("Order Contribution")
plt.show()
```

```python
# Labels + Percent
plt.pie(df["Orders"][:4],
        labels=["D1","D2","D3","D4"],
        autopct="%1.1f%%")
plt.show()
```

```python
# Exploded Pie
plt.pie(df["Orders"][:4],
        explode=[0,0.1,0,0],
        autopct="%1.1f%%")
plt.show()
```

# **7ï¸âƒ£ SUBPLOTS (plt.subplots())**
**ðŸ”¹ What it is**  -> Multiple plots in one figure

**ðŸ”¹ When to use**  -> Dashboards , Comparison analysis

```python
'''
---Syntax---
fig, ax = plt.subplots(rows, cols)
'''
fig, ax = plt.subplots(1,2, figsize=(12,4))

ax[0].plot(df["Day"], df["Sales"])
ax[0].set_title("Sales")

ax[1].plot(df["Day"], df["Profit"])
ax[1].set_title("Profit")

plt.show()
```

```python
# Access Individual Axes
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

ax[0,0].plot(df["Day"], df["Sales"])
ax[0,1].plot(df["Day"], df["Profit"])
ax[1,0].bar(df["Day"], df["Orders"])
ax[1,1].scatter(df["Orders"], df["Sales"])

plt.tight_layout()
plt.show()
```

# **8ï¸âƒ£ ANNOTATION (plt.annotate())**
**ðŸ”¹ What it is**  -> Adds insight text

**ðŸ”¹ When to use**  -> Explain spikes , Highlight max/min

```python
'''
---Syntax---
plt.annotate(text, xy, xytext, arrowprops)
'''
plt.plot(df["Day"], df["Sales"])
plt.annotate("Peak Sales",
             xy=(10,300),
             xytext=(6,250),
             arrowprops=dict())
plt.show()
```

# ðŸ” Two Ways to Plot in Matplotlib
1ï¸âƒ£ Procedural (State-based) â€” plt. style **bold text**

```python
x = df['Day']
y = df['Sales']
plt.plot(x, y)
plt.title("Sales")
plt.xlabel("Day")
plt.show()

'''
Matplotlib internally keeps track of:
   â€œWhich figure is active right now?â€
    - This is implicit (hidden).
'''
```

**2ï¸âƒ£ Object-Oriented â€” fig, ax style**

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Sales")
ax.set_xlabel("Day")
plt.show()
```

# ***What are these objects?***
**ðŸ”¹ Figure**  (fig)->The entire canvas , Think: the whole page / window

**ðŸ”¹ Axes** (ax)-> The actual plotting area , Where lines, bars, points are drawn

**âš ï¸ Important:**

Axes â‰  Axis

Axes â†’ plotting area

Axis â†’ x-axis or y-axis line
