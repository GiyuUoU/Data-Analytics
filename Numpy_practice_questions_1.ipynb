```python
import numpy as np

# Columns: [Student_ID, Study_Hours, Math_Score, Science_Score, English_Score]
data = np.array([
    [1, 2, 65, 70, 68],
    [2, 4, 78, 82, 80],
    [3, 1, 55, 60, 58],
    [4, 5, 88, 90, 85],
    [5, 3, 72, 75, 70],
    [6, 6, 92, 95, 91],
    [7, 2, 60, 64, 62],
    [8, 4, 80, 85, 83],
    [9, 3, 74, 78, 76],
    [10, 5, 89, 92, 90]
])

data.shape
```

```python
'''
1️⃣ Extract Study Hours
Using NumPy slicing, extract only the study hours column.
'''
study_hours = data[: , 1]
study_hours
```

```python
'''
2️⃣ Average Score per Subject
Calculate the average score for Math, Science, and English.
'''
avg_score = data.mean(axis= 0)
avg_score[2:5]
```

```python
'''
3️⃣ Find Top Performer
Find the student with the highest total score across all subjects.
'''
total_marks = data[:, 2:].sum(axis=1)

top_student_index = np.argmax(total_marks)

topper_row = data[top_student_index]
topper_id = data[top_student_index, 0]

print(f"Topper ID: {topper_id}, Total Marks: {total_marks[top_student_index]}")
```

```python
'''
4️⃣ Students Who Studied More Than 3 Hours
Filter students who studied more than 3 hours per day.
'''
filtered_stu = data[0:-1][data[0:-1, 1] > 3]
filtered_stu[0: , 0:2]
```

```python
'''
6️⃣ Students Scoring Above 85 in Math
Find student IDs where Math score > 85.
'''
math_topper = data[data[: , 2] > 85 ,0]
math_topper
```

```python
'''
7️⃣ Add a New Column
Add a new column representing Average Score per Student.
'''
avg_score = data[ : , 2:5 ].mean(axis=1)
rounded_avg = np.round(avg_score , 2)
np.column_stack((data , rounded_avg))
```

```python
'''
8️⃣ Standard Deviation of Scores
Calculate standard deviation for each subject.
'''
subjects_std = data[: , 2:].std(axis=0)
np.round(subjects_std , 2)
```

```python
'''
9️⃣ Normalize Scores
Normalize all subject scores using Min-Max normalization.
'''
scores = data[:, 2:]

col_min = scores.min(axis=0)
col_max = scores.max(axis=0)

normalized_scores = (scores - col_min) / (col_max - col_min)

normalized_scores = np.round(normalized_scores, 2)

print(normalized_scores)
```

```python
'''
🔟 Correlation Between Study Hours and Math Score
Compute correlation between:
'''
study_hours = data[:,1]
math_scores = data[:,2]

corr_matrix = np.corrcoef(study_hours, math_scores)

corr = corr_matrix[0,1]
print(np.round(corr , 2))
```
