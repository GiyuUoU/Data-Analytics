```python
# .value_counts()
df = pd.DataFrame({
    "City": ["Delhi", "Mumbai", "Delhi", "Lucknow", "Mumbai"]
})

df["City"].value_counts()
```

```python
# Using .apply() function -> Can be applied in both Seires and DataFrame
# Meaning - Apply function on rows or columns.
df2 = pd.DataFrame({
    "Salary": [10000, 20000, 30000]
})

df2["Bonus"] = df2["Salary"].apply(lambda x : x*1.5)

df2["Total"] = df2["Salary"] + df2["Bonus"]
df2
```

```python
# Using .map() function => Can only be applied on Series data
# Meaning - Element-wise transformation (only for Series)
df3 = pd.DataFrame({
    "Marks": [50, 70, 90 , 20 , 56 , 67 , 48 , 71 , 66 , 78 , 92 , 83]
})

df3["Grade"] = df3["Marks"].map(lambda x : "Pass" if x >= 60 else "Fail" )
df3
```

```python
# Using .pivot() function
# Meaning - pivot() is used to restructure data (change its shape)
# no duplicate handling and aggregation function support like pivot_table

df4 = pd.DataFrame({
    "Date": ["Jan", "Jan", "Feb", "Feb"],
    "Product": ["A", "B", "A", "B"],
    "Sales": [100, 200, 150, 250]
})
df4
```

```python
df4.pivot( index = "Date" , columns="Product" , values= "Sales")

# Limitation (IMPORTANT) - > Pivot fails if duplicate values exist
```

```python
# Solution - pivot_table()
pd.pivot_table(df4, index="Date", columns="Product", values="Sales", aggfunc="mean")
```

```python
# pivot_table Example  - 1
# Handles duplicate data by giving there mean
# Aggregation functions can be applied on it


df = pd.DataFrame({
    "Month": ["Jan", "Jan", "Feb", "Feb"],
    "Product": ["A", "B", "A", "B"],
    "Sales": [100, 200, 150, 250]
})

pd.pivot_table(df, index="Month", columns="Product", values="Sales")
```

```python
# pivot_table Example  - 2
df = pd.DataFrame({
    "Month": ["Jan", "Jan", "Feb", "Feb"],
    "City": ["Delhi", "Mumbai", "Delhi", "Mumbai"],
    "Sales": [100, 200, 150, 250]
})

pd.pivot_table(df, index=["Month", "City"], values="Sales")
```

```python
# pivot_table Example - 3
pd.pivot_table(df, index="Month", columns="City", values="Sales", aggfunc="sum", margins=True)
```

```python
# Date-Time Functions
df = pd.DataFrame({
    "date": ["2024-01-01", "2024-02-15"]
})

df["date"] = pd.to_datetime(df["date"])

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
df["day"] = df["date"].dt.day

df
```
